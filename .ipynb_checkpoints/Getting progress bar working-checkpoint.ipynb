```python
from tqdm.notebook import tqdm
```

```python
for i in tqdm(range(100)):
    sleep(3)
```

```python
def powers(x):
    for i in tqdm(range(100)):
        
        record = []
        i = 0
        while i < x**3:

            record.append(x**i)
    
    
            i += 1
        
    return record
```

```python
powers(18)
```
