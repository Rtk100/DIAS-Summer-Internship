```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import time 
```

```python
def commutator(A,B):
    return np.dot(A,B) - np.dot(B,A)

def F1(X1,X2):
    return (g**2) * commutator(X2,commutator(X1,X2))

def F2(X1,X2):
    return (g**2) * commutator(X1,commutator(X2,X1))

def H(X1,X2,P1,P2):
    T = 1/2 * np.trace(np.dot(V1,V1) + np.dot(V2,V2))
    U = -(g**2)/4 * np.trace(
        np.dot(commutator(X1,X2),commutator(X1,X2)) + np.dot(commutator(X2,X1),commutator(X2,X1))
    )
    return np.abs(T + U)
```

# With Complex matrices

```python
N = 2

X1 = np.zeros([N,N],dtype = complex)
X2 = np.zeros([N,N],dtype = complex)

for i in np.arange(N):
    
    for j in np.arange(N):
    
        k = np.random.rand()
        m = np.random.rand()
        X1[i][j] = complex(k,m)
        X1[j][i] = complex(k,-m)
        
    if (i == N-1) & (i == j):
        currentsum = complex(0,0)
        for s in np.arange(N-1):
            currentsum += X1[s][s]
        
        X1[i][j] = -currentsum
        
            
for i in range(N):
    
    for j in range(N):

        k = np.random.rand()
        m = np.random.rand()
        X2[i][j] = complex(k,m)
        X2[j][i] = complex(k,-m)
        
    if (i == N-1) & (i == j):
        currentsum = complex(0,0)
        for s in np.arange(N-1):
            currentsum += X2[s][s]
        
        X2[i][j] = -currentsum
        
V1 = np.zeros([N,N], dtype = complex)
V2 = np.zeros([N,N], dtype = complex)

X = np.array([X1,X2])
V = np.array([V1,V2])
```

```python
X1_sol, X2_sol = np.zeros([2,int(1e6),N,N],dtype = complex)
V1_sol, V2_sol = np.zeros([2,int(1e6),N,N],dtype = complex)
F1_sol, F2_sol = np.zeros([2,int(1e6),N,N],dtype = complex)
Energy = np.zeros(int(1e6))

dt = 1e-4
g = 0.001

for i in np.arange(int(1000000)):
    
    F1_sol[i] = F1(X1,X2)
    F2_sol[i] = F2(X1,X2)
    
    X1_sol[i] = X1
    X2_sol[i] = X2
    V1_sol[i] = V1
    V2_sol[i] = V2
    
    X1_n = X1 + V1*dt + F1(X1,X2)*(dt**2)/2
    X2_n = X2 + V2*dt + F2(X1,X2)*(dt**2)/2
    
    V1 = V1 + (F1(X1,X2) + F1(X1_n,X2_n))*dt/2
    V2 = V2 + (F2(X1,X2) + F2(X1_n,X2_n))*dt/2
    
    X1, X2 = X1_n, X2_n
    
    Energy[i] = H(X1,X2,V1,V2)
    print(H(X1,X2,V1,V2))
    
    #if i % 100000 == 0:
     #   print(H(X1,X2,V1,V2))
        
    if (np.isnan(H(X1,X2,V1,V2))):
        print(i)
        break
```

```python
L2_X1 = np.zeros(i)
L2_V1 = np.zeros(i)

L2_X2 = np.zeros(i)
L2_V2 = np.zeros(i)

L2_F1 = np.zeros(i)
L2_F2 = np.zeros(i)

Hamiltonians = np.zeros(i)

for i in np.arange(i):
    
    L2_X1[i] = np.linalg.norm(X1_sol[i])
    L2_V1[i] = np.linalg.norm(V1_sol[i])
    
    L2_X2[i] = np.linalg.norm(X2_sol[i])
    L2_V2[i] = np.linalg.norm(V2_sol[i])
    
    L2_F1[i] = np.linalg.norm(F1_sol[i])
    L2_F2[i] = np.linalg.norm(F2_sol[i])
    
    Hamiltonians[i] = np.linalg.norm(Energy[i])
    
t = np.linspace(0,i*1e-4,i+1)
```

# With Real matrices

```python
N = 2

X1 = np.zeros([N,N])
X2 = np.zeros([N,N])

for i in np.arange(N):
    
    for j in np.arange(N):
    
        k = np.random.rand()
        X1[i][j] = k+5500
        X1[j][i] = k+5500
        
    if (i == N-1) & (i == j):
        currentsum = 0
        for s in np.arange(N-1):
            currentsum += X1[s][s]
        
        X1[i][j] = -currentsum
        
            
for i in range(N):
    
    for j in range(N):

        k = np.random.rand()
        X2[i][j] = k + 5500
        X2[j][i] = k + 5500
        
    if (i == N-1) & (i == j):
        currentsum = 0
        for s in np.arange(N-1):
            currentsum += X2[s][s]
        
        X2[i][j] = -currentsum
        
V1 = np.zeros([N,N])
V2 = np.zeros([N,N])

X = np.array([X1,X2])
V = np.array([V1,V2])
```

```python
X1_sol, X2_sol = np.zeros([2,int(1e5),N,N])
V1_sol, V2_sol = np.zeros([2,int(1e5),N,N])
F1_sol, F2_sol = np.zeros([2,int(1e5),N,N])
Energy = np.zeros(int(1e5))

dt = 1e-4
g = 0.001

for i in np.arange(int(100000)):
    
    F1_sol[i] = F1(X1,X2)
    F2_sol[i] = F2(X1,X2)
    
    X1_sol[i] = X1
    X2_sol[i] = X2
    V1_sol[i] = V1
    V2_sol[i] = V2
    
    X1_n = X1 + V1*dt + F1(X1,X2)*(dt**2)/2
    X2_n = X2 + V2*dt + F2(X1,X2)*(dt**2)/2
    
    V1 = V1 + (F1(X1,X2) + F1(X1_n,X2_n))*dt/2
    V2 = V2 + (F2(X1,X2) + F2(X1_n,X2_n))*dt/2
    
    X1, X2 = X1_n, X2_n
    
    Energy[i] = H(X1,X2,V1,V2)
    print(H(X1,X2,V1,V2))
    
    #if i % 100000 == 0:
     #   print(H(X1,X2,V1,V2))
        
    if (np.isnan(H(X1,X2,V1,V2))):
        print(i)
        break
k = i
```

```python
L2_X1 = np.zeros(int((k+1)/1000))
L2_X2 = np.zeros(int((k+1)/1000))
L2_V1 = np.zeros(int((k+1)/1000))
L2_V2 = np.zeros(int((k+1)/1000))
L2_F1 = np.zeros(int((k+1)/1000))
L2_F2 = np.zeros(int((k+1)/1000))
Hamiltonians = np.zeros(int((k+1)/1000))

for i in np.arange(k+1):

    if i % 1000 == 0:
        
        L2_X1[int(i/1000)] = np.linalg.norm(X1_sol[i])
        L2_X2[int(i/1000)] = np.linalg.norm(X2_sol[i])
        L2_V1[int(i/1000)] = np.linalg.norm(V1_sol[i])
        L2_V2[int(i/1000)] = np.linalg.norm(V2_sol[i])
        
        L2_F1[int(i/1000)] = np.linalg.norm(F1_sol[i])
        L2_F2[int(i/1000)] = np.linalg.norm(F2_sol[i])
        
        Hamiltonians[int(i/1000)]= Energy[i]

t = np.linspace(0,k*1e-4,int((k+1)/1000))
```
